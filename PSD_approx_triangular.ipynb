```python
%matplotlib inline
from pylab import *
import wavfun 
import showfunlab5
import pamfunlab5
```

```python
FB=100
Fs=44100
zn = randn(5*FB)
an = sign(zn)
tt,st1= pamfunlab5.pam11(an,FB,Fs,'rect')   #generating rectangular PAM
st11 = st1**2                               # squaring signal
showfunlab5.showpsd0(st11,Fs,[-1000,1000,0],Fs) #Plot PSD
```

The resulting waveform will be like DC so the PSD will be f = 0

```python
rdt1 = diff(hstack((0,st1)))*Fs                                    # Derivative of st1 
showfunlab5.showpsd0(rdt1,Fs,[-200,200,-25],Fs) #Plot PSD
```

Spectral components at +-FB/2

```python
tt,st2= pamfunlab5.pam11(an,FB,Fs,'tri')   #generating triangular PAM
st22=st2**2                                # squaring signal
showfunlab5.showpsd0(st22,Fs,[-200,200,-60],Fs) #Plot PSD
```

The PSD of triangular PAM squared will give spectral components at FB

```python
rdt2 = diff(hstack((0,st2)))*Fs                                    # Derivative of rt 
showfunlab5.showpsd0(rdt2,Fs,[-200,200,-25],Fs) #Plot PSD
```

```python
tt,st3= pamfunlab5.pam11(an,FB,Fs,'sinc',[20,0])   #generating sinc PAM
st33= st3**2                                        # squaring signal
showfunlab5.showpsd0(st33,Fs,[-200,200,-60],Fs) #Plot PSD
```

The PSD of the sinc pulse squared shows maximum power percentage within limits -FB and +FB (-100 and 100)

```python
rdt3 = diff(hstack((0,st3)))*Fs                                    # Derivative of rt 
showfunlab5.showpsd0(rdt3,Fs,[-200,200,-25],Fs)                 #Plot PSD
```

```python
tt,st4= pamfunlab5.pam11(an,FB,Fs,'rcf',[5,0.5])   #generating RCf PAM
st44 = st4**2                                          # squaring signal
showfunlab5.showpsd0(st44,Fs,[-200,200,-60],Fs) #Plot PSD
```

It is clear from the above PSD of rcf PAM squared that it has spectral components at FB(100)

```python
rdt4 = diff(hstack((0,st4)))*Fs                                    # Derivative of rt 
showfunlab5.showpsd0(rdt4,Fs,[-200,200,-25],Fs) #Plot PSD
```

```python
rt1,Fs = wavfun.wavread('pamsig401.wav')         # reading wav file
xt1 = rt1**2                                       # squaring signal
```

```python
showfunlab5.showpsd0(xt1,Fs,[-400,400,-60],Fs) #Plot PSD
```

Spectral component at f = 360 indicates the FB of this signal

```python
xt1 = rt1**4
showfunlab5.showpsd0(xt1,Fs,[-400,400,-60],Fs) #Plot PSD
```

Spectral component at f = 360 indicates the FB of this signal
