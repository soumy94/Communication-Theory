```python
%matplotlib inline
from pylab import *
import pamfunlab5
import numpy as np
import showfunlab5

Fs = 44100 # Sampling rate 
f1 = 700 # Test frequency 1 
f2 = 720 # Test frequency 1 
tlen = 2 # Duration in seconds 
tt = arange(round(tlen*Fs))/float(Fs) # Time axis 
x1t = sin(2*pi*f1*tt) # Sine with freq f1 
x2t = 0.01*cos(2*pi*f2*tt) # Attenuated cosine with freq f2 
xt = x1t+x2t # Combined sinusoidal signal 
showfunlab5.showpsd0(xt,Fs,[-1000,1000,0],Fs) #Plot S_x(f)
```

showpsd function was added to the showfun module

```python
showfunlab5.showpsd0(xt,Fs,[-1000,1000,-60],Fs) #Plot S_x(f) in dB
```

The plot of Sx(f) in dB is plotted

# (c)Generate about 5 sec of random binary polar PAM signals with rectangular, triangular, and sinc (tail length ≈ 20, no windowing) pulses. Use FB = 100 and Fs = 44100 Hz and display the PSDs of the three signals in dB, with lower limit -60 dB, over the frequency range −500...500 Hz. Set the blocklength to Fs. Verify that the PSDs look right. A reasonable rule of thumb for the bandwidth of PAM signals is 2FB when p(t) is rectangular, FB when p(t) is triangular and FB/2 when p(t) is a sinc pulse. Determine how much of the total signal power is contained within these bandwidths for the rectangular, triangular and sinc pulse cases.


```python

FB=100
Fs=44100
zn = randn(5*FB)
an = sign(zn)
tt,st= pamfunlab5.pam11(an,FB,Fs,'rect')   #generating rectangular PAM
```

```python
showfunlab5.showpsd0(st,Fs,[-200,200,-60],Fs) #Plot PSD
```

The Power Spectral Density of rectangular PAM is plotted. Px(f1,f2) = 95 % of total power is contained in 2FB for the rectangular pulse.

```python
tt,st= pamfunlab5.pam11(an,FB,Fs,'tri')      #generating triangular PAM
```

```python
showfunlab5.showpsd0(st,Fs,[-100,100,-60],Fs) #Plot PSD
```

The Power Spectral Density of triangle PAM is plotted. Around 99.6% of total power is contained in this bandwidth FB of triangualar pulse.

```python
tt,st= pamfunlab5.pam11(an,FB,Fs,'sinc',[20,0])         #generating sinc PAM
```

```python
showfunlab5.showpsd0(st,Fs,[-50,50,-60],Fs) #Plot PSD
```

The Power Spectral Density of sinc PAM is plotted. Around 99.2 % of total power is contained in Bandwidth FB/2 for sinc pulse.
