```python
%matplotlib inline
from pylab import *
import filtfun 
import showfunlab6
import wavfun as wf
```

```python
Fs = 16000
xt = hstack((zeros(Fs/2),1,zeros(Fs/2-1))) #generating an impulse
tt = arange(-0.5*Fs,0.5*Fs)/float(Fs)
yt,n = filtfun.trapfilt(xt,Fs,1000,20,0.2)
```

```python
xlim(-0.005,0.005)
title('Impulse response of filter')
xlabel('Time(t)')
ylabel('Amplitude')
plot(tt,yt)
```

```python
showfunlab6.showft(tt,yt,[-2000,2000,0])          # Absolute value of S(f) , llim >=0
```

```python
showfunlab6.showft(tt,yt,[-2000,2000,-80])             # Normalized value of S(f) , llim < 0
```

# How do the results change when you set  = 0?

```python
yt1,n = filtfun.trapfilt(xt,Fs,1000,20,0)
xlim(-0.005,0.005)
title('Impulse response of filter at alpha = 0')
xlabel('Time(t)')
ylabel('Amplitude')
plot(tt,yt1)
```

# At alpha = 0 , the trapezoidal pulse transforms into sinc pulse.

```python
showfunlab6.showft(tt,yt1,[-2000,2000,0])          # llim >= 0
```

Fourier transform on linear scale (absolute)

```python
showfunlab6.showft(tt,yt1,[-2000,2000,-40])         #llim < 0 
```

Fourier transform in dB

# For alpha = 0 , it will be a sinc pulse and we will get a rectangular waveform in fft from -fL to fL (Bandwidth is 2fL)

```python
rt,Fs = wf.wavread('pamsig601.wav')
```

```python
showfunlab6.showpsd0(rt,Fs,[-500,500,0],Fs)
```

```python
yt2,n = filtfun.trapfilt(rt,Fs,500,10,0.2)
yt22 = yt2*yt2
```

```python
showfunlab6.showpsd0(yt22,Fs,[-820,820,-25],Fs)
```

From the above PSD plot we can deduce that FB = 815 Hz

```python
showfunlab6.showeye(yt2,Fs,815,100,[0,3,-8000,8000])
```

For fL = 500 the eye diagram is optimum. Below this effect of ISI can be seen in the eye diagram.
fL is nearly equal to (1+alpha)FB/2 optimum for bandlimiting the signal.
